### Which method of plotting is better? (Pyplot or matplotlib OO method)
- When Plotting something quickly, Okay to use the pyplot method
- When plotting something more advanced, use the OO method

```python
Over_Fifty = Heart_disease[Heart_disease["age"] > 50]
Over_Fifty
```

```python
# Pyplot method
Over_Fifty.plot(x="age", 
               y="thalach",
               kind="scatter",
               c="target",
               figsize=(10, 6));
```

```python
# oo method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
Over_Fifty.plot(kind="scatter",
               x="age",
               y="chol",
               c="target",
               ax=ax);
ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter  = ax.scatter(x=Over_Fifty["age"],
                     y=Over_Fifty["chol"],
                     c=Over_Fifty["target"])

# Customize the plot
ax.set(title="Heart Diesease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line(Mean line)
ax.axhline(Over_Fifty["chol"].mean(),
          linestyle="--");
```

```python
Over_Fifty.head()
```

```python
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                       ncols=1,
                       figsize=(10, 10),
                              sharex=True);# this to share the same x values and label between 2 or more x axises
ax0.scatter(x=Over_Fifty["age"],
                    y=Over_Fifty["chol"],
                    c=Over_Fifty["target"]);
ax0.set(title="Heart disease and Cholesterol Levels",
       ylabel="Cholesterol");
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(y=Over_Fifty["chol"].mean(),
           linestyle="--");
ax1.scatter(x=Over_Fifty["age"],
           y=Over_Fifty["thalach"],
           c=Over_Fifty["target"]);
ax1.set(title="Heart disease and Max heart Rate",
       xlabel="Age",
       ylabel="Max heart rate");
ax1.legend(*scatter.legend_elements(), title="Target");
ax1.axhline(y=Over_Fifty["thalach"].mean(),
           linestyle="--");

# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                       ncols=1,
                       figsize=(10, 10),
                              sharex=True);# this to share the same x values and label between 2 or more x axises
ax0.scatter(x=Over_Fifty["age"],
                    y=Over_Fifty["chol"],
                    c=Over_Fifty["target"],
           cmap="winter");
ax0.set(title="Heart disease and Cholesterol Levels",
       ylabel="Cholesterol");
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(y=Over_Fifty["chol"].mean(),
           linestyle="--");
ax0.set_xlim([50, 80])
ax1.scatter(x=Over_Fifty["age"],
           y=Over_Fifty["thalach"],
           c=Over_Fifty["target"],
           cmap="winter");
ax1.set(title="Heart disease and Max heart Rate",
       xlabel="Age",
       ylabel="Max heart rate");
ax1.legend(*scatter.legend_elements(), title="Target");
ax1.axhline(y=Over_Fifty["thalach"].mean(),
           linestyle="--");
ax1.set_ylim([60, 200])
# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
fig.savefig("Heart-disease-analysis.png")
```
